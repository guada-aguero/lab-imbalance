# Inbalanced Classes
## In this lab, we are going to explore a case of imbalanced classes. 


Like we disussed in class, when we have noisy data, if we are not careful, we can end up fitting our model to the noise in the data and not the 'signal'-- the factors that actually determine the outcome. This is called overfitting, and results in good results in training, and in bad results when the model is applied to real data. Similarly, we could have a model that is too simplistic to accurately model the signal. This produces a model that doesnt work well (ever).

```python
# Sample the data
fraud_sample = fraud.sample(n=100000, random_state=42)

# Verify the sample
print(fraud_sample.shape)
display(fraud_sample.head())
```

### What is the distribution of the outcome?

### Clean the dataset. Pre-process it to make it suitable for ML training. Feel free to explore, drop, encode, transform, etc. Whatever you feel will improve the model score.

```python
# Your code here
fraud_sample.info()
```

```python
# Drop columns that are not useful for the model
fraud_sample = fraud_sample.drop(columns=['nameOrig','nameDest'])
fraud_sample.head()
```

```python
# Encode Categorical Variables
# One-hot encode the 'type' column
fraud_sample = pd.get_dummies(fraud_sample, columns=['type'], drop_first=True).astype(int)
fraud_sample.head()
```

```python
# Let's create additional features that could be useful for fraud detection.

# Feature engineering: creating balance differences

fraud_sample['balanceOrig_diff'] = fraud_sample['oldbalanceOrg'] - fraud_sample['newbalanceOrig']
fraud_sample['balanceDest_diff'] = fraud_sample['oldbalanceDest'] - fraud_sample['newbalanceDest']

# Flagging zero balances
fraud_sample['flagOrig_zero_balance'] = fraud_sample['oldbalanceOrg'].apply(lambda x:1 if x == 0 else 0)
fraud_sample['flagNew_zero_balance'] = fraud_sample['newbalanceOrig'].apply(lambda x:1 if x == 0 else 0)
fraud_sample['flagDest_zero_balance'] = fraud_sample['oldbalanceDest'].apply(lambda x:1 if x == 0 else 0)
fraud_sample['flagNewDest_zero_balance'] = fraud_sample['newbalanceDest'].apply(lambda x:1 if x == 0 else 0)

fraud_sample.head()
```

```python
# Scale Numerical Features
# List of numerical columns to scale
numerical_cols = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'balanceOrig_diff', 'balanceDest_diff']

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler and transform the data
fraud_sample[numerical_cols] = scaler.fit_transform(fraud_sample[numerical_cols])
```

```python
# Final check on the processed dataset
print(fraud_sample.info())
print(fraud_sample.head())
```

### Run a logisitc regression classifier and evaluate its accuracy.

```python
# Define the features (X) and the target variable (y)

X = fraud_sample.drop('isFraud', axis=1)
y = fraud_sample['isFraud']

# Split the data into training (70%) and testing (30%) sets

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

# Check the shape of the resulting datasets
print(f"Training set: X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"Test set: X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
# Initialize the Logistic Regression model

model = LogisticRegression(max_iter=1000, random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)
```

```python
# Predict on the test set

y_pred = model.predict(X_test)
```

```python
# Evaluate the Model's Performance

# Accuracy

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy: .4f}')

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Generate the Classification report

class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)
```

```python
"""
1. Accuracy
Accuracy: 0.9995 (or 99.95%):
This indicates that the model correctly predicted 99.95% of the instances in the test set. 
However, accuracy alone can be misleading in imbalanced datasets like this. This high accuracy is largely driven by the model's success in predicting the majority class (non-fraudulent transactions).

2. Confusion Matrix

[[29958     2]
 [   14    26]]
True Negatives (TN): 29,958: Non-fraudulent transactions correctly classified.
False Positives (FP): 2: Non-fraudulent transactions incorrectly classified as fraudulent.
False Negatives (FN): 14: Fraudulent transactions incorrectly classified as non-fraudulent.
True Positives (TP): 26: Fraudulent transactions correctly classified.

3. Classification Report
Precision (for Class 1 - Fraudulent) = 0.93:
Out of all the transactions predicted as fraudulent, 93% were actually fraudulent. This is quite high, indicating the model is good at minimizing false positives.

Recall (for Class 1 - Fraudulent) = 0.65:
The model correctly identified 65% of all actual fraudulent transactions. This shows that the model misses 35% of fraudulent transactions, which might be critical in a real-world scenario.


F1-Score (for Class 1 - Fraudulent) = 0.76:
The F1-score is the harmonic mean of precision and recall. This score indicates a balance between precision and recall, but the recall could still be improved to catch more fraudulent transactions.

"""
```

### Now pick a model of your choice and evaluate its accuracy.

```python
# Train a Random Forest Model

# Initialize the Random Forest model

rf_model = RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced')

# Train the model on the training data

rf_model.fit(X_train, y_train)
```

```python
# Predict on the test set

y_pred_rf = rf_model.predict(X_test)
```

```python
# Evaluate the model's accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")

# Generate the confusion matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Random Forest Confusion Matrix:")
print(conf_matrix_rf)

# Generate the classification report
class_report_rf = classification_report(y_test, y_pred_rf)
print("Random Forest Classification Report:")
print(class_report_rf)
```

```python
# I can't really see a huge improvemnt by simply using the class_weight so I'll attempt to optimize the Hyperparameters
# Define the hyperparameter grid

param_grid = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}

# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42, class_weight='balanced')

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_grid,
    n_iter=100,  # Number of parameter settings that are sampled
    cv=3,  # 3-fold cross-validation
    verbose=2,  # Controls the verbosity: the higher, the more messages
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fit the random search model
random_search.fit(X_train, y_train)

# Get the best parameters
best_params = random_search.best_params_
print("Best Parameters:", best_params)
```

```python
# Initialize the Random Forest model with the best parameters
best_rf_model = RandomForestClassifier(
    n_estimators=500,
    min_samples_split=5,
    min_samples_leaf=4,
    max_features='log2',
    max_depth=10,
    bootstrap=True,
    random_state=42,
    class_weight='balanced'
)

# Train the model on the training data
best_rf_model.fit(X_train, y_train)

# Predict on the test set
y_pred_best_rf = best_rf_model.predict(X_test)

# Evaluate the model's accuracy
accuracy_best_rf = accuracy_score(y_test, y_pred_best_rf)
print(f"Optimized Random Forest Accuracy: {accuracy_best_rf:.4f}")

# Generate the confusion matrix
conf_matrix_best_rf = confusion_matrix(y_test, y_pred_best_rf)
print("Optimized Random Forest Confusion Matrix:")
print(conf_matrix_best_rf)

# Generate a classification report
class_report_best_rf = classification_report(y_test, y_pred_best_rf)
print("Optimized Random Forest Classification Report:")
print(class_report_best_rf)
```
